```python
import random
import csv
```

```python
stock = "AMD"
path_stock = f"Stocks/{stock}.csv"

stock_array = []

with open(path_stock, mode='r', newline='') as stock:
    stock_reader = csv.reader(stock)
    while(True):
        stock_reader = next(csv.reader(stock))[1]
        if stock_reader == "--":
            break
      
        stock_array.append(round(float(stock_reader) , 2))
                
                
             
print(stock_array)

def sum(arraya, start, end):
    sum = 0
    for i in range(end-start):
        sum += arraya[start+i]
    return sum

```

```python
# randomized buy and sells
x = 200 # amout of days for one full test
total = 100 # some starting value
# 3 actions of buy sell hold for each day
stock = "AMD"
day1 = len(stock_array)

day_bought = 0  # when on array of last buy
hold = 0  # if currently holding a stock
sum_total = []
action = 0

for day in range(x):
    # get value of the day 1

    if hold == 1: # allready have the stock
        action = random.randint(0,2)  # sell or hold
        if action == 1: # sell 
            total += total * (sum(stock_array, day_bought, day)/100)
            day_bought = - 1
            hold = 0
        else:
            total += total*(stock_array[day]/100) 
    else: 
        # hold = 0 npt
        action = random.randint(0,2)  # Buy or wait
        if action == 1: # sell
            hold = 1
            day_bought = day  
        total += total*(stock_array[day]/100)     
    sum_total.append(total)  # only see the value change when selling 
# if hold == 1:
#     total += total * (sum(stock_array, day_bought, day)/100)
    

print(total)
print(stock_array)
print("from start to end", 100 + sum(stock_array, 0, len(stock_array)))  # need to change 100 if use other value for starting value * 100 / 100
```

```python
# graphing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(sum_total, label=f'Total $ over {len(sum_total)} days')
plt.title(f'Total $ over {len(sum_total)} Days')
plt.xlabel('Days')
plt.ylabel('Total')
plt.legend()
plt.grid(True)
plt.show()
```
